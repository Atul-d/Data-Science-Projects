**Speechiness --> -->** Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

 - Based on the above description we have to filter the record or do binning in range 0 to 0.33 to 0.66 to 1

```python
def binning(bin):
    if bin >= 0.33:
        return 0
    if 0.34 <= bin <= 0.66:
        return 1
    else:
        return 2

vfunc = np.vectorize(binning)

df['speechiness_bin'] = vfunc(df['speechiness'])
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
model_lr = classifier.fit(X_train, Y_train)
```

```python
Y_pred = model_lr.predict(X_test)
print(list(zip(Y_test, Y_pred)))
```

```python
print(df.columns[:-1])
print("*" * 75)

print(classifier.coef_)
print("*" * 75)

print(classifier.coef_.ravel())
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import f1_score
from sklearn import metrics

cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc = accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

## Ensemble Model

```python
# we will use 4 different models for training
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, log_loss, accuracy_score
from sklearn.datasets import make_classification

models_ensemble = {
    'LogisticRegression': LogisticRegression(random_state = 42),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SVC': SVC(random_state = 42),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=42)
}
```

```python
def loss(Y_test, Y_pred, retu=False):
    pre = precision_score(Y_test, Y_pred)
    rec = recall_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)
    loss = log_loss(Y_test, Y_pred)
    acc = accuracy_score(Y_test, Y_pred)
    
    if retu:
        return pre, rec, f1, loss, acc
    else:
        print('  pre: %.3f\n  rec: %.3f\n  f1: %.3f\n  loss: %.3f\n  acc: %.3f' % (pre, rec, f1, loss, acc))
# train_eval_train

def train_eval_train(models_ensemble, X, y):
    for name, model in models_ensemble.items():
        print(name,':')
        model.fit(X, y)
        loss(y, model.predict(X))
        print('-'*30)
        
train_eval_train(models_ensemble, X_train, Y_train)
```
