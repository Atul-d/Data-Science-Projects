```python
# Most of the independent variable are not normally distributed and hence we will transform it into Normally Distribution using Scaling method.
# Perform Scalling into the X variable.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
# After performing Transform it gets converted to Series so we have to convert it to DataFrame
X = pd.DataFrame(X)
X
```

```python
# Because of the conversion from series to dataFrame column name is in form of 0,1,2,3 and we need to rename it.
# Just extract column from main data i.e. car.columns and then copy all the variable instead of Y i.e. price and paste it.
X.columns = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration',
       'num-of-doors', 'body-style', 'drive-wheels', 'engine-location',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg']
X
```

```python
# Find Correlation and heat map of the Independent variable
corr_df=X.corr(method="pearson")
print(corr_df)
plt.figure(figsize = (15,7))
sns.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True)
```

```python
# VIF is used to check for multi-collinearity in the data

# Package used to Find VIF. VIF value greater than 5 means mulitcollinearity is present.
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
# Created an empty data frame i.e. vif_df
vif_df = pd.DataFrame()
# All the X variable is being saved in new column called features in vif_df data frame.
vif_df["features"] = X.columns
# Created another column VIF Factor in vif_df data frame and after = code means ki calculate vif for all the rows in i data.
# i data means ki range(X.shape[1]) this code means ki find the range of column of X variable. i.e. 0-2 hoga
# In first iteration it will run for i = 0 i.e. TV (Y variable) find R^2 uske baad for i = 1 for Radio(Y) and so on
# Detailed explanation in Notebook
vif_df["VIF Factor"] = [vif(X.values, i) for i in range(X.shape[1])]
# Format the value upto 2 decimal
vif_df.round(2)
# jiska VIF greater than 5 remove one by one after dropping one variable run VIF again and check 5 se jyada hai kay if yes do this process again
# TO drop column use drop() function
```

```python
# There are certain variable having High VIF but first we will run a complete model and then we will do tuning.
```

```python
# Package required to split the data.
from sklearn.model_selection import train_test_split
# Splitting the data into train and test
# If test_size is 0.2 toh it will split the data into 80% Train and 20% Test data
# If test_size is 0.3 toh it will split the data into 70% Train and 30% Test data
# by default test_size = 0.25
# for more than 1000 observation use 0.3 and for less than 1000 use 0.2 as test_size value.
# X_train contains 80% of the data from X_var(TV, radio, newspaper) and X_test contains 20% of the Data from X_var(TV, radio, newspaper)
# Y_train contains 80% of the data from Y_var(Sales) and Y_test contains 20% of the Data from Y_var(Sales)
# X, y are dependent and independet variable so X-train, y_train has 80% data and X_test and Y_test has 20% data.
# random_state is like set.seed() in R
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_log, test_size=0.2,
                                                    random_state=10)  
# Reason we are taking Y_log is because Y was not normally distributed and hence we transformed it.
```

```python
# Running Linear regression on training model
from sklearn.linear_model import LinearRegression
#create a model object
lm = LinearRegression()
#train the model object. X_Train has all the value from X _var(Radio, TV, Newspaper) and Y_train contain value form Sales column.
# Use for training part of the model
lm.fit(X_train,Y_train)

# print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)
```

```python
# pair the feature names with the coefficients 
# Below code will display column_name and Coefficient value side by side
print(list(zip(X.columns, lm.coef_)))
```

```python
# Predict function is done on X-Axis.
# PRedicting the model using predict()
Y_pred = lm.predict(X_test)
Y_pred
```

```python
new_df=pd.DataFrame()
new_df=X_test
# Basically comparing what we have predicted(Actual Sales) and what system has predicted(Predicted Sales).
new_df["Actual sales"]=Y_test
new_df["Predicted sales"]=Y_pred
new_df
```

```python
# To find R^2, Adj R^2 and RMSE used with Y-Actual and Y-pred variable
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(Y_test,Y_pred)
print(r2)

rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print(rmse)

adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print(adjusted_r_squared)
```

```python
# To check wdr RMSE value is good or not
print(min(Y_test))
print(max(Y_test))
```
